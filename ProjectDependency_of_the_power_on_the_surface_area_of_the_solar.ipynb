```python
# دالة الطباعة الملونة والعريضة
def bold_blue(text):
    return f"\033[1;34m{text}\033[0m"  # أزرق وغامق

def bold_green(text):
    return f"\033[1;32m{text}\033[0m"  # أخضر وغامق

def bold_cyan(text):
    return f"\033[1;36m{text}\033[0m"  # سماوي وغامق

def bold_magenta(text):
    return f"\033[1;35m{text}\033[0m"  # بنفسجي وغامق

# --------------------------------------------------
print(bold_blue("EQUATIONS USED IN CALCULATIONS:"))
# --------------------------------------------------
print("Power Equation:")
print("P = Uoc × Isc")
print()

# --------------------------------------------------
print(bold_green("DATA OF TABLES:"))
from tabulate import tabulate
import pandas as pd

# البيانات
Area = [
    "1 (Without Shading)",
    "3/4 (Shading of 1/4)",
    "1/2 (Shading of 1/2)",
    "1/4 (Shading of 3/4)"
]

Uoc = [0.530, 0.517, 0.504, 0.480]
Isc = [32.98, 27.20, 19.08, 10.56]
Power = [round(Uoc[i] * Isc[i], 3) for i in range(len(Uoc))]

# تحويلها إلى DataFrame
df = pd.DataFrame({
    "Active Area": Area,
    "Uoc (V)": Uoc,
    "Isc (mA)": Isc,
    "Power (mW)": Power
})

# طباعة الجدول باستخدام tabulate 
print("\033[1;36m------ Table of Results ------\033[0m")
print(tabulate(df, headers='keys', tablefmt='fancy_grid', showindex=False))

import matplotlib.pyplot as plt
import numpy as np

# بياناتك الأصلية
areas = np.array([1.0, 0.75, 0.5, 0.25])
Uoc = np.array([0.530, 0.517, 0.504, 0.480])
Isc = np.array([32.98, 27.20, 19.08, 10.56])
Power = np.array([17.48, 14.07, 9.62, 5.07])

# نبدأ الرسم
plt.figure(figsize=(15, 4))

# --------------------------------------
# Uoc vs Area
plt.subplot(1, 3, 1)
plt.scatter(areas, Uoc, color='purple', label='Data Points')

# رسم خط مستقيم
uoc_fit = np.poly1d(np.polyfit(areas, Uoc, 1))
x_line = np.linspace(0.2, 1.05, 100)
plt.plot(x_line, uoc_fit(x_line), color='gray', linestyle='-', label='Best Fit Line')

plt.title("Uoc (V) vs Active Area")
plt.xlabel("Active Area")
plt.ylabel("Uoc (V)")
plt.grid(True)
plt.legend()

# --------------------------------------
# Isc vs Area
plt.subplot(1, 3, 2)
plt.scatter(areas, Isc, color='teal', label='Data Points')

isc_fit = np.poly1d(np.polyfit(areas, Isc, 1))
plt.plot(x_line, isc_fit(x_line), color='gray', linestyle='-', label='Best Fit Line')

plt.title("Isc (mA) vs Active Area")
plt.xlabel("Active Area")
plt.ylabel("Isc (mA)")
plt.grid(True)
plt.legend()

# --------------------------------------
# Power vs Area
plt.subplot(1, 3, 3)
plt.scatter(areas, Power, color='orange', label='Data Points')

pwr_fit = np.poly1d(np.polyfit(areas, Power, 1))
plt.plot(x_line, pwr_fit(x_line), color='gray', linestyle='-', label='Best Fit Line')

plt.title("Power (mW) vs Active Area")
plt.xlabel("Active Area")
plt.ylabel("Power (mW)")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()



# --------------------------------------------------
print(bold_magenta("DISCUSSION & ANALYSIS OF RESULTS:"))
# --------------------------------------------------
print("- As the active surface area of the solar cell decreases due to shading,")
print("  • The open-circuit voltage (Uoc) drops gradually.")
print("  • The short-circuit current (Isc) drops significantly.")
print("  • Therefore, the output power decreases noticeably.")
print()

# --------------------------------------------------
print(bold_green("METHODS TO REDUCE EFFECT OF SHADING IN PVs:"))
# --------------------------------------------------
print("- Use bypass diodes to allow current to flow around shaded cells.")
print("- Arrange solar panels to avoid shading from nearby objects.")
print("- Keep modules clean and unobstructed to ensure maximum exposure.")
print()

# --------------------------------------------------
print(bold_blue("EXPERIMENTAL ERRORS:"))
# --------------------------------------------------
volt_error = float(input("Enter the smallest division on the voltmeter: "))
amp_error = float(input("Enter the smallest division on the ammeter: "))

v_accurcy = volt_error * 0.5
i_accurcy = amp_error * 0.5

print(f"Voltmeter accurcy = ±{v_accurcy} V")
print(f"Ammeter accurcy = ±{i_accurcy} mA")
```

```python
import tkinter as tk
from PIL import Image, ImageDraw, ImageTk

# القيم الخاصة بالتجربة حسب نسبة التظليل
Uoc_values = [0.530, 0.517, 0.504, 0.480]
Isc_values = [32.98, 27.20, 19.08, 10.56]
Power_values = [round(Uoc_values[i] * Isc_values[i], 2) for i in range(4)]

# دالة لإنشاء صورة خلية شمسية مع تظليل حسب النسبة
def create_solar_image(shading_percent):
    img_width, img_height = 400, 300
    solar_img = Image.new("RGB", (img_width, img_height), "blue")
    draw = ImageDraw.Draw(solar_img)
    for y in range(1, 6):
        draw.line([(0, y * 50), (img_width, y * 50)], fill="black", width=2)
    for x in range(1, 8):
        draw.line([(x * 50, 0), (x * 50, img_height)], fill="black", width=2)
    shade_height = int((shading_percent / 100) * img_height)
    draw.rectangle([(0, 0), (img_width, shade_height)], fill=(100, 100, 100))
    return solar_img
# دالة تحديث الصورة والقيم
def update_simulation(val):
    shade = int(val)
    index = int(shade / 25)
    img = create_solar_image(shade)
    tk_img = ImageTk.PhotoImage(img)
    image_label.config(image=tk_img)
    image_label.image = tk_img
    uoc = Uoc_values[index]
    isc = Isc_values[index]
    power = Power_values[index]
    uoc_label.config(text=f"Uoc: {uoc} V")
    isc_label.config(text=f"Isc: {isc} mA")
    power_label.config(text=f"Power: {power} mW")
    shading_label.config(text=f"Shading: {shade}%")

# إنشاء واجهة التطبيق
root = tk.Tk()
root.title("Interactive Solar Cell Simulation")
root.geometry("450x500")
root.resizable(False, False)

# العنوان
title = tk.Label(root, text="Solar Cell Simulation", font=("Helvetica", 16, "bold"))
title.pack(pady=10)
# شريط التظليل
shading_label = tk.Label(root, text="Shading: 0%", font=("Helvetica", 12))
shading_label.pack()
slider = tk.Scale(root, from_=0, to=75, orient="horizontal", resolution=25,
                  length=300, command=update_simulation)
slider.pack()

# عرض الصورة
initial_img = create_solar_image(0)
tk_initial_img = ImageTk.PhotoImage(initial_img)
image_label = tk.Label(root, image=tk_initial_img)
image_label.image = tk_initial_img
image_label.pack(pady=10)

# عرض القيم
uoc_label = tk.Label(root, text="Uoc: 0.530 V", font=("Helvetica", 12))
uoc_label.pack()
isc_label = tk.Label(root, text="Isc: 32.98 mA", font=("Helvetica", 12))
isc_label.pack()
power_label = tk.Label(root, text="Power: 17.48 mW", font=("Helvetica", 12, "bold"))
power_label.pack()

# تشغيل التطبيق
root.mainloop()
```
